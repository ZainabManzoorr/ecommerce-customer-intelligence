```python
df.dropna(inplace=True)
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
for x in df.index:
  if df.loc[x,'Quantity'] < 0:
    df.drop(x, inplace=True)
```

```python
pd.get_dummies(df,columns=['Country'],prefix='Country')
```

```python
junk_codes = ['POST', 'D', 'M', 'TEST', 'BANK', 'CRUK','A','B','D','E','G']
df = df[~df['StockCode'].isin(junk_codes)]
```

```python
df = df[df['Quantity'] < 100000]
```

```python
df = df[df['UnitPrice'] >= 0]
```
