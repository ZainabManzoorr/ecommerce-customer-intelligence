```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
transactions = pd.read_csv("transactions.csv")
```

```python
transactions['InvoiceDate'] = pd.to_datetime(transactions['InvoiceDate'])

plt.figure(figsize=(12,6))
transactions.groupby(transactions['InvoiceDate'].dt.date)['Revenue'].sum().plot()
plt.title("Daily Revenue Over Time")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=transactions['Hour'],y=transactions['Revenue'],estimator=sum,errorbar=None)
plt.title("Revenue by hour of Day")
plt.xlabel('Hour')
plt.ylabel("Total Revenue")
plt.show()
```

```python
sns.boxplot(x='WeekDay',y='Revenue',data=transactions)
plt.title("Revenue Distribution by Weekday")
plt.xlabel("Weekdau(0=Monday)")
plt.ylabel("Revenue")
plt.show()
```
